```python
from bokeh.plotting import figure, show, output_notebook

output_notebook()

p = figure(title="Line Chart", x_axis_label='X', y_axis_label='Y')
p.line([1, 2, 3, 4], [4, 7, 2, 5], line_width=2)
show(p)
```

```python
from bokeh.plotting import figure, output_notebook, show
output_notebook()

p = figure(
    title="Learning Bokeh Plot",         
    x_axis_label='X Axis',               
    y_axis_label='Y Axis',              
    width=600,                          
    height=400                           
)
p.line([1, 2, 3, 4, 5], [3, 7, 8, 5, 1], line_width=2)
show(p)
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import Legend
output_notebook()

p = figure(
    title="Customized Bokeh Plot",
    x_axis_label="X Axis Label",
    y_axis_label="Y Axis Label",
    x_range=(0, 10),    
    y_range=(0, 20),     
    width=600,
    height=400,
    background_fill_color="#f5f5f5",  
    tools="pan,box_zoom,reset,save" 
)
line1 = p.line([1, 2, 3, 4], [4, 6, 5, 9],
               line_width=2,
               color="blue",
               legend_label="Line A")

circle1 = p.circle([1, 2, 3, 4], [4, 6, 5, 9],
                   size=8,
                   color="red",
                   alpha=0.8,
                   legend_label="Points A")

p.legend.location = "top_left"
p.legend.click_policy = "hide"  
show(p)
```
